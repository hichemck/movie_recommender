```python
import pandas as pd
import numpy as np
```

```python
ratings_df = pd.read_csv('../data/ml-latest-small/ratings.csv')
ratings_df_sorted = ratings_df.sort_values(['userId','movieId'])
ratings_df_sorted.tail()
```

```python
user_item_matrix = ratings_df_sorted.pivot(index='userId', columns='movieId', values='rating')
user_item_matrix
```

## Random recommender

```python
movies_df = pd.read_csv('../data/ml-latest-small/movies.csv')
```

```python
sorted_rating_sums = user_item_matrix.sum(axis=0).sort_values(ascending=False)
sorted_rating_sums
```

```python
proba_series = sorted_rating_sums / sorted_rating_sums.sum()
proba_series
```

```python
def recommend_randomly():
    recommended_movie_id = np.random.choice(proba_series.index, p=proba_series.values)
    recommended_movie = movies_df[movies_df['movieId']==recommended_movie_id]['title'].values[0]
    return recommended_movie_id, recommended_movie
```

```python
recommend_randomly()
```

```python
# sum of unseen films for each user
unseen_sum_vs_user = user_item_matrix.T.isna().sum().sort_values(ascending=False)
unseen_sum_vs_user
```

```python
user_correlation_matrix = user_item_matrix.T.corr()
user_correlation_matrix.tail()
```

```python
def get_similar_users(user_id):
    user_correlation = user_correlation_matrix.iloc[user_correlation_matrix.index==user_id].T.sort_values(by=user_id, \
                                                        ascending=False).dropna()
    best_similar_users = user_correlation[user_correlation>0.2].dropna()[1:]
    return best_similar_users.index
```

```python
users_not_seen_films = user_item_matrix.T.isna()
users_not_seen_films.tail()
```

```python
users_seen_films = user_item_matrix.T.notna()
users_seen_films.tail()
```

```python
def get_unseen_films(user_id):
    user_boolean_series = users_not_seen_films[[user_id]]
    not_seen_films = user_boolean_series.index[user_boolean_series[user_id]].tolist()
    return not_seen_films
```

```python
def get_seen_films(user_id):
    user_seen_boolean_series = users_seen_films[[user_id]]
    seen_films = user_seen_boolean_series.index[user_seen_boolean_series[user_id]].tolist()
    return seen_films
```

```python
def get_sorted_films_by_rating(user_id):
    output = user_item_matrix.loc[[user_id]].T.sort_values(by=user_id, ascending=False).dropna().index
    return output.tolist()
```

```python
def suggest_best_film_from_best_similar_user(user_id) : 
    
    seen_films = get_seen_films(user_id)    
    sorted_similar_users = get_similar_users(user_id)  
    
    next_user = sorted_similar_users[0]
    suggestion = get_sorted_films_by_rating(next_user)[0]    
    j = 0 
    i = 0
    
    while suggestion in seen_films:
        i+=1
        suggestion = get_sorted_films_by_rating(next_user)[i]
        if i == len(get_sorted_films_by_rating(next_user)) -1 :            
            j+=1
            next_user = sorted_similar_users[j]
            i = -1
    return suggestion
```
